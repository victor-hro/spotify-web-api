```python
from openai import OpenAI
import os
import requests
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO
from dotenv import find_dotenv, load_dotenv

load_dotenv(find_dotenv())

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),
)

response = client.images.generate(
    prompt='Homer simpson as a computer science teacher',
    n=1,
    style='natural',
    size="512x512",
    quality="standard",
)

img_url = response.data[0].url
img_response = requests.get(img_url)

# Carregar a imagem usando PIL
imagem = Image.open(BytesIO(img_response.content))

# Exibir a imagem usando Matplotlib
plt.imshow(imagem)
plt.axis('off')  # Para remover os eixos (opcional)
plt.show()
```

```python
import streamlit as st
from openai import OpenAI
import os
import requests
from PIL import Image
from io import BytesIO
from dotenv import find_dotenv, load_dotenv

load_dotenv(find_dotenv())

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),
)

st.title("Image Generation App")

prompt = st.text_input("Enter prompt", 'Homer simpson as a computer science teacher')

if st.button('Generate Image'):
    response = client.images.generate(
        prompt=prompt,
        n=1,
        style='natural',
        size="512x512",
        quality="standard",
    )

    img_url = response.data[0].url
    img_response = requests.get(img_url)

    # Carregar a imagem usando PIL
    imagem = Image.open(BytesIO(img_response.content))

    # Exibir a imagem
    st.image(imagem, caption='Generated Image', use_column_width=True)
```
